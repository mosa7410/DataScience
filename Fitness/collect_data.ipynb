```python
import json
import os
import pandas as pd
import numpy as np

folder_list = os.listdir('./sokulee')[1:-3]

for f in folder_list :
    
    path = './sokulee/' + f
    file_list = os.listdir(path)
    
    heart_list = []
    sleep_list = []
    steps_list = []

    for i in file_list :
        if i.find('heart') is not -1 :
            heart_list.append(i)
        if i.find('sleep') is not -1 :
            sleep_list.append(i)
        if i.find('steps') is not -1 :
            steps_list.append(i)

    file_path = path + '/'

    for file_name in heart_list :
        with open(file_path + file_name) as json_file :
            data_list = []

            json_data = json.load(json_file)
            json_list = list(json_data.keys())
            if 'activities-heart' in json_list :
                json_date = json_data['activities-heart']
                date = json_date[0]['dateTime']

                heart_data = json_data['activities-heart-intraday']['dataset']
                if heart_data != [] :
                    for num in range(len(heart_data)) :
                        data_list.append([f, date , heart_data[num]['time'] , heart_data[num]['value']])

                else :
                    data_list.append([f, date ,'',''])
               
            else :
                date = file_name.split('_')[1]
                data_list.append([f, str(date[:4]) + '-' + str(date[4:6]) + '-' + str(date[6:]),'',''])
                 
            data_list = pd.DataFrame(data_list)
            data_list.to_csv('./sokulee/heart/' + f + date + '.csv', index = False, header=False)
            
    for file_name in sleep_list :
        with open(file_path + file_name) as json_file :
            data_list = []

            json_data = json.load(json_file)
            json_list = list(json_data.keys())
            if 'sleep' in json_list :
                json_data = json_data['sleep']

                if json_data != [] :
                    date = json_data[0]['dateOfSleep']
                    start_time = json_data[0]['startTime'].split('T')[1]
                    sleep_data = json_data[0]['minutesAsleep']
                    wake_data = json_data[0]['minutesAwake']
                    bed_data = json_data[0]['timeInBed']

                    data_list.append([f, date, start_time, sleep_data , wake_data, bed_data])

                else :
                    date = file_name.split('_')[1]
                    data_list.append([f, str(date[:4]) + '-' + str(date[4:6]) + '-' + str(date[6:]), '', '', '', ''])
                    
            else :    
                date = file_name.split('_')[1]
                data_list.append([f, str(date[:4]) + '-' + str(date[4:6]) + '-' + str(date[6:]), '', '', '', ''])
                
            data_list = pd.DataFrame(data_list)
            data_list.to_csv('./sokulee/sleep/' + f + date + '.csv', index = False, header=False)

    for file_name in steps_list :
        with open(file_path + file_name) as json_file :
            data_list = []

            json_data = json.load(json_file)
            json_list = list(json_data.keys())
            if 'activities-steps' in json_list :
                json_date = json_data['activities-steps']
                date = json_date[0]['dateTime']

                steps_data = json_data['activities-steps-intraday']['dataset']

                if steps_data != [] :
                    for num in range(len(steps_data)) :
                        sp = steps_data[num]['time'].split(':')
                        data_list.append([f, date, steps_data[num]['time'], sp[0], sp[1],steps_data[num]['value']])

                else :
                    data_list.append([f,date,'','','',''])
            else :
                date = file_name.split('_')[1]
                data_list.append([f, str(date[:4]) + '-' + str(date[4:6]) + '-' + str(date[6:]),'','','',''])
                    
            data_list = pd.DataFrame(data_list)
            data_list.to_csv('./sokulee/steps/' + f + date + '.csv', index = False, header=False)
```
