```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
heart = pd.read_csv('./data/heart.csv', names = ['user','date','time', 'heartrate'])
heart.head(3)
```

```python
sleep = pd.read_csv('./data/sleep.csv', names = ['user', 'date', 'startTime', 'minutesAsleep', 'minutesAwake', 'timeInbed'])
sleep.head(3)
```

```python
steps = pd.read_csv('./data/steps.csv', names = ['user', 'date', 'time', 'hour', 'minute', 'steps'])
steps.head(3)
```

```python
weather = pd.read_csv('./data/weather.csv')
weather.head(3)
```

```python
heart.fillna(0)
mean_heart = heart.groupby(['user','date'])['heartrate'].mean().reset_index()
mean_heart.head(3)
```

```python
mean_steps = steps.groupby(['user', 'date'])['steps'].sum().reset_index()
steps = mean_steps['steps'].div(2)
mean_steps['steps'] = steps
mean_steps.head(3)
```

```python
all_heart = mean_heart.groupby(['date'])['heartrate'].mean().reset_index()
all_sleep = sleep.groupby(['date'])['minutesAsleep'].mean().reset_index()
all_steps = mean_steps.groupby(['date'])['steps'].mean().reset_index()
```

```python
total_data = weather.merge(all_heart, how = 'left', on = 'date')
total_data = total_data.merge(all_sleep, how = 'left', on = 'date')
total_data = total_data.merge(all_steps, how = 'left', on = 'date')
total_data.head(3)
```

```python
total_data.to_csv('./data/relation.csv', index = False)
```
